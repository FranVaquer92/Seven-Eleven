```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


def seven_eleven():
    dado1 = np.random.choice(6, 1)+1          #Lanzamos el dado 1
    dado2 = np.random.choice(6, 1)+1          #Lanzamos el dado 2
    dados = dado1 + dado2                     #La suma de ambos dados
    if dados == 7 or dados == 11:             #Si el resultado es 7 u 11, ganas
        resultado = 1
    elif dados == 2 or dados == 3 or dados == 12:     #Si el resultado es 2, 3 o 12 pierdes
        resultado = 0
    else:                                     #Si no se da ninguno de los resultados mencionados anteriormente se vuelve a lanzar
        numero = dados
        dado1 = np.random.choice(6, 1)+1
        dado2 = np.random.choice(6, 1)+1
        dados = dado1 + dado2
        while dados != numero:                #Si repites el resultado obtenido anteriormente, ganas
            if dados == 7:                    #Si sacas un 7 pierdes
                resultado = 0
                break
            dado1 = np.random.choice(6, 1)+1
            dado2 = np.random.choice(6, 1)+1
            dados = dado1 + dado2
        if numero == dados:
            resultado = 1
    return resultado
```

```python
N = 1000000
partidas = [seven_eleven() for i in range(N)]
ganadas = np.sum(partidas)
perdidas = N - ganadas
probabilidad_ganar = ganadas/N
```

```python
print('Ganadas: ', ganadas)
print('Perdidas ', perdidas)
print(f'La probabilidad de ganar es de un {probabilidad_ganar*100:.2f}%')
print(probabilidad_ganar)
#print(np.std(partidas)/np.sqrt(N))
plt.plot(np.cumsum(partidas)/(np.arange(N)+1))
plt.xlabel('Intentos')
plt.ylabel('% ganados')
plt.title('Porcentaje de partidas ganadas')
plt.show()
print('Para ganar a la primera hay que sacar un 7 o un 11 en la primera tirada. Hay 6 posibilidades de sacar un 7: (1,6),(6,1),(2,5),(5,2),(3,4),(4,3). Por otro lado hay 2 posibilidades de sacar un 11: (5,6),(6,5). Por lo tanto son 8 posibilidades de ganar en la primera tirada de 36 combinaciones totales. De modo que la probabilidad de ganar a la primera es 8/36 = 0,22')
```
